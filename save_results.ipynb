```python
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d  # 
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
import librosa
import subprocess
import shutil
import os
from sklearn.metrics import confusion_matrix
import seaborn as sn
```

```python
data = pd.read_csv('./tests/results.csv')
```

```python
print(data.query('activation_name == "relu"').sort_values('accuracy',ascending=False).head(1))
print(data.query('activation_name == "tanh"').sort_values('accuracy',ascending=False).head(1))
print(data.query('activation_name == "sigmoid"').sort_values('accuracy',ascending=False).head(1))
```

```python

def save_results(path,final_path,name,y):
    y_pred = np.load(path + 'pred.npy')
    testing_error = np.load(path + 'error_val.npy')
    training_error = np.load(path + 'error_train.npy')
    cm = confusion_matrix(y,y_pred,normalize='true')

    if not os.path.exists(final_path): 
        os.mkdir(final_path)

    # print(cm)
    plt.figure(figsize=(12,8))
    plt.title(f'Confusion matrix for {name} best result')
    sn.heatmap(cm,annot=True,cmap='Blues',cbar=False)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.savefig(final_path + 'confusion_matrix')
    # plt.clf()
    plt.close()

    plt.title(f'Errors for {name} best result')
    plt.plot([i for i in range(len(training_error))],training_error,label='training')
    plt.plot([i for i in range(len(testing_error))],testing_error,label='validation')
    plt.xlabel('Epoch')
    plt.ylabel('Error')
    plt.legend()
    plt.savefig(final_path + 'error_plot')
    # plt.clf()
    plt.close()




    args = ['cp',path + '*.csv',final_path]
    subprocess.run(' '.join(args),shell=True)


    
    
```

```python

test_file = './datasets/sound_class_test.csv'
y = pd.read_csv(test_file).to_numpy().argmax(axis=1)
results_dirs = [('./tests/relu_50___0.01/','relu'),('./tests/tanh_200_100_50_0.01/','tanh'),('./tests/sigmoid_50___0.15/','sigmoid')]


for result_dir,name in results_dirs:
    save_results(result_dir,f'./best_results/{name}/' ,name,y)
```
